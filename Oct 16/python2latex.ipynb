```python
import numpy as np
import pandas as pd
```

Let's import some data ... say some of the data used in the Gaia hack.

```python
df = pd.read_csv('James_Data.csv')
df.head(5)
```

Latex doesn't like underscores in column names so let's replace those first.

```python
df.columns
df.columns = [x.replace('_', ' ') for x in df.columns]
#df.columns = [''.join(('$', str(i), '$')) for i in df.columns]
df.columns
```

What if we wanted to do some fancy and a bit more complicated, like having errors in the same column i.e. $1.4^{+0.1}_{-0.1}$ ?

```python
df.head(5)
```

```python
def proper_errors(x):
    x['proper errors'] = '$'+str(x['astero parallax'])+'^{+'+str(x['astero parallax err'])+'}'+'_{-'+str(x['astero parallax err'])+'}$'
    return x
df = df.round(2)
df_errors = df.apply(proper_errors, axis=1)
```

```python
df_errors.head()
```

```python
print df_errors.to_latex(index=False, escape=False)
```
