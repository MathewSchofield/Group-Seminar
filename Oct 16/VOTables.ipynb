```python
from astropy.io.votable import parse
import numpy as np
import xml.etree.ElementTree as ET
import pandas as pd
```

## VOTable's the hard way

Let's try and parse the file as an xml file (which it should be!)

```python
tree = ET.parse('result.xml')
```

```python
root = tree.getroot()
root
```

```python
for node in tree.iter():
    print(node.text)
```

## The easy way

```python
# Parse votable
votable = parse("result.vot")

# Get table from votable
table = votable.get_first_table()

# Convert to numpy rec array
data = table.array

# Convert rec array to pandas dataframe
df = pd.DataFrame.from_records(data)

df.head()
```
